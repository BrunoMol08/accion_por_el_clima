# Python script para hacer un melt en los datos

## Librerías importadas

```python
import pandas as pd
```

### Footprint

```python
df = pd.read_csv('/home/bruno/Documentos/Visualizacion_informacion/proyecto_vizualizaciones_13/Footprint_pcap_tonnes.csv')
```

```python
df.head()
```

```python
df_correction = pd.melt(df, id_vars=['country'], value_vars=df.columns.values[2:],
    var_name ='year', value_name ='footprint')
```

```python
df_correction
```

```python
df_correction.to_csv('/home/bruno/Documentos/Visualizacion_informacion/proyecto_vizualizaciones_13/Footprint_pcap_tonnes_with_melt.csv',
                        index=False)
```
